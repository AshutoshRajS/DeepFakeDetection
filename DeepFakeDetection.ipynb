```python
import torch
import timm
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.models as models
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, roc_curve
import numpy as np

# Device Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define dataset directories
DataDir = '/content/drive/MyDrive/ExtractedFace'
train_dir = DataDir
val_dir   = DataDir
test_dir  = DataDir

# Image Preprocessing
img_size = (224, 224)
batch_size = 32

transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = ImageFolder(root=train_dir, transform=transform)
val_dataset = ImageFolder(root=val_dir, transform=transform)
test_dataset = ImageFolder(root=test_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(train_dataset.class_to_idx)

# Define Model Wrapper
class ModelWrapper(nn.Module):
    def __init__(self, base_model, input_size):
        super(ModelWrapper, self).__init__()
        self.base = base_model
        self.fc = nn.Sequential(
            nn.Linear(input_size, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.base(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

# Load Pretrained Models
resnet50 = timm.create_model("resnet50", pretrained=True, num_classes=0)
xception = timm.create_model("xception", pretrained=True, num_classes=0)
vgg16 = models.vgg16(pretrained=True)

# Freeze Feature Extractors
for model in [resnet50, xception, vgg16]:
    for param in model.parameters():
        param.requires_grad = False

# Create Individual Models
model_resnet50 = ModelWrapper(resnet50, 2048).to(device)
model_xception = ModelWrapper(xception, 2048).to(device)
model_vgg16 = ModelWrapper(vgg16.classifier[:-1], 4096).to(device)

# Define Hybrid Model
class HybridModel(nn.Module):
    def __init__(self, base1, base2):
        super(HybridModel, self).__init__()
        self.base1 = base1
        self.base2 = base2
        self.fc = nn.Sequential(
            nn.Linear(2048 + 2048, 512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x1 = self.base1(x)
        x2 = self.base2(x)
        x = torch.cat((x1.view(x1.size(0), -1), x2.view(x2.size(0), -1)), dim=1)
        return self.fc(x)

model_hybrid = HybridModel(resnet50, xception).to(device)

# Loss & Optimizers
criterion = nn.BCELoss()
optimizers = {
    "ResNet50": optim.Adam(model_resnet50.fc.parameters(), lr=0.0001),
    "Xception": optim.Adam(model_xception.fc.parameters(), lr=0.0001),
    "VGG16": optim.Adam(model_vgg16.fc.parameters(), lr=0.0001),
    "Hybrid": optim.Adam(model_hybrid.fc.parameters(), lr=0.0001)
}

# Training Function
def train_model(model, train_loader, val_loader, optimizer, num_epochs=50,save_path="model.pth"):
    history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}

    best_val_acc = 0  # Track the best validation accuracy

    for epoch in range(num_epochs):
        model.train()
        running_loss, correct, total = 0.0, 0, 0

        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            images, labels = images.to(device), labels.float().to(device).view(-1, 1)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            predicted = (outputs > 0.5).float()
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

        train_acc = correct / total
        val_acc, val_loss = evaluate_model(model, val_loader, return_loss=True)

        history["train_loss"].append(running_loss / len(train_loader))
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f},Val loss: {val_loss:.4f}")

        # Save the model if it achieves better validation accuracy
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), save_path)
            print(f"Best model saved at epoch {epoch+1} with val_acc: {val_acc:.4f}")

    return history

# Evaluation Function
def evaluate_model(model, test_loader, return_loss=False):
    model.eval()
    correct, total = 0, 0
    predictions, true_labels = [], []
    running_loss = 0.0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.float().to(device).view(-1, 1)
            outputs = model(images)
            loss = criterion(outputs, labels)

            predicted = (outputs > 0.5).float()
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

            predictions.extend(outputs.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

            running_loss += loss.item()

    acc = accuracy_score(true_labels, (np.array(predictions) > 0.5).astype(int))
    precision = precision_score(true_labels, (np.array(predictions) > 0.5).astype(int))
    recall = recall_score(true_labels, (np.array(predictions) > 0.5).astype(int))
    f1 = f1_score(true_labels, (np.array(predictions) > 0.5).astype(int))
    auc = roc_auc_score(true_labels, np.array(predictions))

    if return_loss:
        return acc, running_loss / len(test_loader)

    return acc, precision, recall, f1, auc, classification_report(true_labels, (np.array(predictions) > 0.5).astype(int))

# Train All Models
print("Training ResNet50...")
history_resnet50 = train_model(model_resnet50, train_loader, val_loader, optimizers["ResNet50"], save_path="resnet50_best.pth")

print("Training Xception...")
history_xception = train_model(model_xception, train_loader, val_loader, optimizers["Xception"], save_path="xception_best.pth")

#print("Training VGG16...")
#history_vgg16 = train_model(model_vgg16, train_loader, val_loader, optimizers["VGG16"], save_path="vgg16_best.pth")

print("Training Hybrid Model...")
history_hybrid = train_model(model_hybrid, train_loader, val_loader, optimizers["Hybrid"], save_path="hybrid_best.pth")

# Evaluate All Models
metrics_resnet50 = evaluate_model(model_resnet50, test_loader)
metrics_xception = evaluate_model(model_xception, test_loader)
#metrics_vgg16 = evaluate_model(model_vgg16, test_loader)
metrics_hybrid = evaluate_model(model_hybrid, test_loader)

print("ResNet50 Metrics:", metrics_resnet50)
print("Xception Metrics:", metrics_xception)
#print("VGG16 Metrics:", metrics_vgg16)
print("Hybrid Model Metrics:", metrics_hybrid)


import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report

def plot_model_performance(models, histories, test_loader, device, class_labels):
    """
    Plots Training & Validation Loss/Accuracy Curves, AUC-ROC Curves, and Confusion Matrices for multiple models.

    Parameters:
    - models (dict): Dictionary of model names and corresponding PyTorch models.
    - histories (dict): Dictionary of model names and their training histories.
    - test_loader (DataLoader): DataLoader for test dataset.
    - device (torch.device): Device to run model inference.
    - class_labels (list): List of class names for the confusion matrix.
    """
    num_models = len(models)
    fig, axes = plt.subplots(num_models, 4, figsize=(20, 5 * num_models))

    for i, (model_name, model) in enumerate(models.items()):
        history = histories[model_name]
        epochs = range(1, len(history["train_loss"]) + 1)

        # Loss Curve
        axes[i, 0].plot(epochs, history["train_loss"], label="Train Loss", color='blue')
        axes[i, 0].plot(epochs, history["val_loss"], label="Val Loss", color='red')
        axes[i, 0].set_xlabel("Epochs")
        axes[i, 0].set_ylabel("Loss")
        axes[i, 0].set_title(f"{model_name} - Loss Curve")
        axes[i, 0].legend()
        axes[i, 0].grid(True)

        # Accuracy Curve
        axes[i, 1].plot(epochs, history["train_acc"], label="Train Acc", color='blue')
        axes[i, 1].plot(epochs, history["val_acc"], label="Val Acc", color='red')
        axes[i, 1].set_xlabel("Epochs")
        axes[i, 1].set_ylabel("Accuracy")
        axes[i, 1].set_title(f"{model_name} - Accuracy Curve")
        axes[i, 1].legend()
        axes[i, 1].grid(True)

        # AUC-ROC Curve
        model.eval()
        y_true, y_scores, y_preds = [], [], []
        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)

                y_true.extend(labels.cpu().numpy())
                y_scores.extend(outputs.cpu().numpy())
                y_preds.extend((outputs > 0.5).cpu().numpy())  # Convert to binary predictions

        fpr, tpr, _ = roc_curve(y_true, y_scores)
        roc_auc = auc(fpr, tpr)

        axes[i, 2].plot(fpr, tpr, color="blue", label=f"AUC = {roc_auc:.4f}")
        axes[i, 2].plot([0, 1], [0, 1], linestyle="--", color="gray")
        axes[i, 2].set_xlabel("False Positive Rate")
        axes[i, 2].set_ylabel("True Positive Rate")
        axes[i, 2].set_title(f"{model_name} - AUC-ROC Curve")
        axes[i, 2].legend(loc="lower right")
        axes[i, 2].grid(True)

        # Confusion Matrix
        cm = confusion_matrix(y_true, y_preds)
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels, ax=axes[i, 3])
        axes[i, 3].set_xlabel("Predicted Labels")
        axes[i, 3].set_ylabel("True Labels")
        axes[i, 3].set_title(f"{model_name} - Confusion Matrix")

        # Print Classification Report
        print(f"\nClassification Report for {model_name}:\n")
        print(classification_report(y_true, y_preds, target_names=class_labels))

    plt.tight_layout()
    plt.show()


# Model Dictionary (Mapping model names to PyTorch models)
models = {
    "ResNet50": model_resnet50,
    "Xception": model_xception,
   # "VGG16": model_vgg16,
    "Hybrid Model": model_hybrid
}

# Training History Dictionary
histories = {
    "ResNet50": history_resnet50,
    "Xception": history_xception,
  #  "VGG16": history_vgg16,
    "Hybrid Model": history_hybrid
}

# Define Class Labels
class_labels =  ['FAKE','REAL'] # Update based on your dataset

# Call Function to Plot All Models
plot_model_performance(models, histories, test_loader, device, class_labels)
```
